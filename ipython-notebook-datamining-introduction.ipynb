```python
# Importer le module pandas
import pandas
```

```python
# Lire la feuille de calcul contenant le jeu de données (médicaments remboursés par l'assurance maladie en France entre 2008 et 2013)
products = pandas.read_excel('MEDICAM 2008-2013-AMELI.xls', sheetname='MedicAM 0813')
```

```python
# Afficher le nombre d'observations et de variables du jeu de données
products.shape
```

```python
# Afficher les dix premières observations du jeu de données
products.head(10)
```

```python
# Afficher le nom et le type de chaque variable
products.dtypes
```

```python
# Afficher des statistiques sur chaque variable
products.describe(include='all')
```

```python
# Afficher le nombre de médicaments par "classe EphMRA", dans l'ordre décroissant
products.groupby('Classe EphMRA').size().order(ascending=False)
```

```python
# Même chose qu'au-dessus en pourcentage du nombre total de médicaments
(products.groupby('Classe EphMRA').size() / products.size).order(ascending=False)
```

```python
# Même chose pour les "codes EphMRA" en utilisant le raccourci 'value_counts()'
products['Code EphMRA'].value_counts()
```

```python
classification = pandas.read_csv('first-level-classification.csv')
```

```python
# Afficher le jeu de données que l'on vient de lire : la correspondance entre la première lettre du code EphMRA et la classe générale
classification
```

```python
# Ajouter une variable ('code_letter') au jeu de données : la première lettre du code EphMRA
products['code_letter'] = products['Code EphMRA'].str[0]
```

```python
# Afficher la variable 'code_letter' que l'on vient d'ajouter au jeu de données
products['code_letter']
```

```python
# Joindre le jeu de données des médicaments à celui des classes (jointure du côté des médicaments)
products = products.merge(classification, how='left', left_on='code_letter', right_on='LETTER')
```

```python
# Compter le nombre de valeurs manquantes dans la variable "CLASS" que l'on vient de générer par jointure
products['CLASS'].isnull().value_counts()
```

```python
# Importer les modules de représentation graphique
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Compter le nombre de médicaments par classe (comme ci-dessus), assigner le résultat dans la variable 'n_products_by_class' au lieu de l'afficher
n_products_by_class = products['CLASS'].value_counts()
```

```python
# Tracer un diagramme à barres horizontales ('barh') pour la série de données 'n_products_by_class'
n_products_by_class.order().plot(kind='barh', figsize=(16,10))
```

```python
# Faire la somme, par classe, des montants remboursés en 2013
reimbursement_amount_by_class = products.groupby('CLASS')['Montant remboursé 2013'].sum()
```

```python
reimbursement_amount_by_class.order().plot(kind='barh', figsize=(16,10), color=sns.xkcd_rgb["pale red"])
```

```python
# Joindre les deux séries de données calculées ci-dessus ('n_products_vs_reimbursement_amount')
n_products_by_class.name = 'n_products_by_class'
reimbursement_amount_by_class.name = 'reimbursement_amount_by_class'

n_products_vs_reimbursement_amount = pandas.concat([n_products_by_class, reimbursement_amount_by_class], axis=1)
```

```python
# Afficher le résultat
n_products_vs_reimbursement_amount_by_class
```

```python
# Trier les classes par montant remboursé décroissant
n_products_vs_reimbursement_amount = n_products_vs_reimbursement_amount.sort('reimbursement_amount_by_class', ascending=False)
# Tracer le montant remboursé et le nombre de médicaments dans chaque classe, sur un même diagramme à barre avec deux échelles en ordonnées
n_products_vs_reimbursement_amount['reimbursement_amount_by_class'].plot(kind='bar', figsize=(16,10), color=sns.xkcd_rgb["pale red"], legend=True)
n_products_vs_reimbursement_amount['n_products_by_class'].plot(kind='bar', position=1, secondary_y=True, legend=True)
```

```python
# Ecrire dans un fichier CSV le montant remboursé et le nombre de médicaments dans chaque classe
n_products_vs_reimbursement_amount.to_csv('n_products_vs_reimbursement_amount_by_class.csv')
```

```python
# Pour chaque classe, tracer dans un diagramme à barres l'évolution des montants remboursés chaque année entre 2008 et 2013
reimbursement_amount_by_class_by_year = products[['CLASS',
                                                  'Montant remboursé 2008',
                                                  'Montant remboursé 2009',
                                                  'Montant remboursé 2010',
                                                  'Montant remboursé 2011',
                                                  'Montant remboursé 2012',
                                                  'Montant remboursé 2013']].groupby('CLASS').sum()

reimbursement_amount_by_class_by_year.plot(kind='bar', figsize=(16,10))
```
